```python
def find_header_row_idx(table):
    for r_idx in range(min(6, len(table.rows))):
        row_text = " | ".join([c.text.strip().lower() for c in table.rows[r_idx].cells])
        if ("study feature" in row_text
            and "hypothetical target trial" in row_text
            and "target trial emulation" in row_text):
            return r_idx
    return None

t = doc.tables[1]
header_row_idx = find_header_row_idx(t)
print("header_row_idx:", header_row_idx)
```

```python
import re

def squash(s):
    return re.sub(r"[^a-z0-9]+", "", (s or "").lower())

def find_table_and_header(doc, max_scan_rows=40):
    for ti, t in enumerate(doc.tables):
        for r_idx in range(min(max_scan_rows, len(t.rows))):
            row_text = " ".join([c.text for c in t.rows[r_idx].cells])
            z = squash(row_text)
            if ("studyfeature" in z and "hypotheticaltargettrial" in z and "targettrialemulation" in z):
                return ti, r_idx
    return None, None

ti, hr = find_table_and_header(doc)
print("found table index:", ti, "header row:", hr)
```

```python
t = doc.tables[0]
hr = 1  # header row index

feature_map = {}
for r in t.rows[hr+1:]:
    cells = [c.text.strip().replace("\n"," ") for c in r.cells]
    if len(cells) < 3:
        continue

    feature = cells[0].strip()
    ideal   = cells[1].strip()
    emul    = cells[2].strip()

    if feature:
        feature_map[feature] = (ideal, emul)

print("n_features:", len(feature_map))
print("keys preview:", list(feature_map.keys())[:15])
```

```python
import re

ELEMENTS_7 = [
    "Eligibility Criteria",
    "Treatment Strategies",
    "Assignment Procedure",
    "Follow-up Period",
    "Outcome",
    "Causal Contrasts",
    "Analysis Plan",
]

def norm(s):
    return re.sub(r"\s+", " ", (s or "").strip())

def get(key):
    return feature_map.get(key, ("", ""))

ref_ideal = {k: "" for k in ELEMENTS_7}
ref_emul  = {k: "" for k in ELEMENTS_7}

# 1 Eligibility
a,b = get("Eligibility")
ref_ideal["Eligibility Criteria"] = a
ref_emul["Eligibility Criteria"]  = b

# 2 Treatment arms -> Treatment Strategies
a,b = get("Treatment arms")
ref_ideal["Treatment Strategies"] = a
ref_emul["Treatment Strategies"]  = b

# 3 Assignment procedure = Treatment assignment + Anchor point/Index date
a1,b1 = get("Treatment assignment")
a2,b2 = get("Anchor point/Index date")
ref_ideal["Assignment Procedure"] = norm(f"{a1} {a2}")
ref_emul["Assignment Procedure"]  = norm(f"{b1} {b2}")

# 4 Follow up -> Follow-up Period
a,b = get("Follow up")
ref_ideal["Follow-up Period"] = a
ref_emul["Follow-up Period"]  = b

# 5 Primary outcome -> Outcome
a,b = get("Primary outcome")
ref_ideal["Outcome"] = a
ref_emul["Outcome"]  = b

# 6 Causal contrast -> Causal Contrasts
a,b = get("Causal contrast")
ref_ideal["Causal Contrasts"] = a
ref_emul["Causal Contrasts"]  = b

# 7 Analysis plan = Covariates + Confounding adjustment + Statistical analysis
a_cov,b_cov = get("Covariates")
a_con,b_con = get("Confounding adjustment")
a_sta,b_sta = get("Statistical analysis")
ref_ideal["Analysis Plan"] = norm(f"{a_cov} {a_con} {a_sta}")
ref_emul["Analysis Plan"]  = norm(f"{b_cov} {b_con} {b_sta}")

# 확인(앞부분만)
for k in ELEMENTS_7:
    print(k, "=>", ref_ideal[k][:120])
```

```python
import csv, io

GEMINI_COL_IDEAL = "(1) Target Trial Specification (Ideal RCT)"
GEMINI_COL_EMUL  = "(2) Emulation Using Observational Data (Real-World)"

gemini_csv_text = r"""Element,(1) Target Trial Specification (Ideal RCT),(2) Emulation Using Observational Data (Real-World)
Eligibility Criteria,"Patients with moderate-to-severe asthma meeting specific biomarker thresholds (e.g., IgE levels or eosinophil counts) and a history of exacerbations.","Patients in the database with an asthma diagnosis, documented use of high-dose ICS-LABA, and specific lab values/ICD codes matching the trial's clinical criteria."
Treatment Strategies,"Arm A: OmalizumabArm B: MepolizumabArm C: DupilumabAll administered at standardized doses.","Initiation of Omalizumab, Mepolizumab, or Dupilumab defined by the first prescription/dispensing date (Index Date) after a period of non-use."
Assignment Procedure,"Participants are randomly assigned to one of the three arms at baseline (concealed allocation).","Patients are ""assigned"" based on the drug they actually received. Propensity Score (PS) weighting or matching is used to balance baseline confounders."
Follow-up Period,"Starts at randomization and ends at the first occurrence of the outcome, death, or end of study (e.g., 52 weeks).","Starts at the initiation of the mAb (Index Date) and ends at the outcome, treatment discontinuation/switch, death, or loss to follow-up."
Outcome,"Rate of severe asthma exacerbations (e.g., requiring systemic steroids or hospitalization) or change in FEV1.","Identified via ICD codes for emergency visits/hospitalizations or new prescriptions for oral corticosteroids (OCS) bursts."
Causal Contrasts,"Intention-to-treat (ITT): Effect of being assigned to a drug. Per-protocol: Effect of staying on the drug as prescribed.","ITT-like: Effect of initiating the drug (ignoring later switches). Per-protocol: Using inverse probability weighting to account for censoring if treatment is stopped."
Analysis Plan,"Comparison of exacerbation rates using Poisson or negative binomial regression across randomized groups.","Marginal structural models or Cox proportional hazards models adjusted for baseline covariates (age, BMI, smoking, baseline OCS use)."
"""

def parse_gemini_csv(text: str):
    f = io.StringIO(text.strip())
    reader = csv.DictReader(f)
    ideal, emul = {}, {}
    for row in reader:
        elem = (row.get("Element") or "").strip()
        if not elem:
            continue
        ideal[elem] = (row.get(GEMINI_COL_IDEAL) or "").strip()
        emul[elem]  = (row.get(GEMINI_COL_EMUL)  or "").strip()
    return ideal, emul

gem_ideal, gem_emul = parse_gemini_csv(gemini_csv_text)
print(gem_ideal.keys())
```

```python
import numpy as np
import torch

@torch.no_grad()
def embed_meanpool(text: str, tokenizer, model, max_length=256):
    text = (text or "").strip()
    if not text:
        return np.zeros((model.config.hidden_size,), dtype=np.float32)

    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        max_length=max_length,
        padding=True
    )
    outputs = model(**inputs)
    last_hidden = outputs.last_hidden_state  # (1, seq, hidden)
    mask = inputs["attention_mask"].unsqueeze(-1)  # (1, seq, 1)
    masked = last_hidden * mask
    summed = masked.sum(dim=1)
    denom = mask.sum(dim=1).clamp(min=1e-9)
    emb = (summed / denom).squeeze(0).cpu().numpy().astype(np.float32)
    return emb

def cos_sim(a, b, eps=1e-9):
    return float(np.dot(a,b) / ((np.linalg.norm(a)+eps) * (np.linalg.norm(b)+eps)))

def sectionwise_scores(ref_dict, cand_dict, tokenizer, model, max_length=256, order=ELEMENTS_7):
    per = {}
    for k in order:
        e1 = embed_meanpool(ref_dict.get(k, ""), tokenizer, model, max_length=max_length)
        e2 = embed_meanpool(cand_dict.get(k, ""), tokenizer, model, max_length=max_length)
        per[k] = cos_sim(e1, e2)
    overall = float(np.mean(list(per.values())))
    return per, overall

# Ideal vs Ideal
per_ideal, overall_ideal = sectionwise_scores(ref_ideal, gem_ideal, tokenizer, model, max_length=256)

# Emulation vs Emulation
per_emul, overall_emul = sectionwise_scores(ref_emul, gem_emul, tokenizer, model, max_length=256)

print("=== Ideal RCT column ===")
for k,v in per_ideal.items():
    print(f"{k:22s} {v:.4f}")
print("Overall:", round(overall_ideal, 4))

print("\n=== Observational emulation column ===")
for k,v in per_emul.items():
    print(f"{k:22s} {v:.4f}")
print("Overall:", round(overall_emul, 4))
```
